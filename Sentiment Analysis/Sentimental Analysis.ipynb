```python
def variance_threshold_select(df, thresh=0.5):
    df1 = df.copy(deep=True) # Make a deep copy of the dataframe
    selector = VarianceThreshold(thresh)
    selector.fit(df1)
    df2 = df.loc[:,selector.get_support(indices=False)] 

    return df2
```
